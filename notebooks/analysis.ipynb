```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# import coffee data
coffee = pd.read_csv("../data/crop/coffee_output.csv")

# import weather data
weather = pd.read_csv("../data/weather/weather_data.csv")
```

```python
# TODO: Merge these two dataframes together

#)
MG_merged = coffee.merge(weather, on=['year', 'subdivision'], how='outer')
MG_merged
```

```python
# TODO: View the first 5 rows of your data
MG_merged.head()
```

```python
# TODO: Check for null values 
MG_merged.isna()
```

```python
# TODO: Drop rows that contain null values

merged_complete = MG_merged.dropna()
merged_complete.isnull().sum()
```

```python
# TODO: BEGIN YOUR EDA!


#Bivariate Pearson Correlation for reationship between 60kg Bags and weather conditions
bivar_data_a = merged_complete[["60kgs_bag","temp_avg","hum_min", "rain_max", "wind_avg"]].corr()
bivar_data_a
```

```python
#Bivariate visualization of the relationship between 60kg Bags and weather conditions

sns.regplot(data = merged_complete, x= "60kgs_bag", y="temp_avg")
plt.xlabel("Number of Bags at 60kg")
plt.ylabel('Average Temperature')
plt.title('60kg Bags vs Average Temperature for Minas Gerais')
plt.show()

sns.regplot(data= merged_complete, x="60kgs_bag", y="hum_min")
plt.xlabel("Number of Bags at 60kg")
plt.ylabel('Minimum Humidity')
plt.title('60kg Bags vs Minimum Humidity for Minas Gerais')
plt.show()

sns.regplot(data = merged_complete, x= "60kgs_bag", y= "rain_max")
plt.xlabel("Number of Bags at 60kg")
plt.ylabel('Maximum Rain')
plt.title('60kg Bags vs Maximum Rain for Minas Gerais')
plt.show()

sns.regplot(data = merged_complete, x= "60kgs_bag", y= "wind_avg")
plt.xlabel("Number of Bags at 60kg")
plt.ylabel('Average Wind')
plt.title('60kg Bags vs Average Wind for Minas Gerais')
plt.show()
```

```python
#Pearson Correlation for the reationship between nonbearing_trees and weather conditions
bivar_data_b = merged_complete[["nonbearing_trees", "temp_avg","hum_min", "rain_max", "wind_avg"]].corr()
bivar_data_b
```

```python
#Bivariate visualization of the relationship between nonbearing_trees and weather conditions
sns.regplot(data = merged_complete, x= "nonbearing_trees", y= "temp_avg")
plt.xlabel("Nonbearing Trees")
plt.ylabel('Average Temperature')
plt.title('Nonbearing Trees vs Average Temperature for Minas Gerais')
plt.show()

sns.regplot(data = merged_complete, x= "nonbearing_trees", y="hum_min")
plt.xlabel("Nonbearing Trees")
plt.ylabel('Minimum Humidity')
plt.title('Nonbearing Trees vs Minimum Humidity for Minas Gerais')
plt.show()

sns.regplot(data= merged_complete, x= "nonbearing_trees", y="rain_max")
plt.xlabel("Nonbearing Trees")
plt.ylabel('Maximum Rain')
plt.title('Nonbearing Trees vs Maximum Rain for Minas Gerais')
plt.show()

sns.regplot(data= merged_complete, x= "nonbearing_trees", y="wind_avg")
plt.xlabel("Nonbearing Trees")
plt.ylabel('Average Wind')
plt.title('Nonbearing Trees vs Average Wind for Minas Gerais')
plt.show()
```
