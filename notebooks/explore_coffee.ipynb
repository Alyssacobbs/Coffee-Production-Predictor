```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
coffee = pd.read_csv("../data/crop/coffee_output.csv")
```

```python
# TODO: view first 5 rows of your data
coffee.head()
```

```python
# TODO: describe your numeric columns
coffee.describe()
```

```python
# TODO: display the null values in your dataframe
coffee.isnull()
```

```python
# TODO: drop your null values
#coffee.isnull().sum()
coffee = coffee.dropna()
coffee.isnull().sum()
```

```python
# TODO: Select only Minas Gerais

coffee_MG = coffee[(coffee.subdivision == "Minas Gerais")]
coffee_MG
```

```python
# TODO: BEGIN YOUR EDA!
#Univarate visualization of Bearing trees across the years using a line plot
coffee_MG.plot(x= 'year', y= 'bearing_trees')
plt.xlabel('Year')
plt.ylabel('Total Trees')
plt.title('Average Bearing Trees Over Years')
plt.grid(True)
plt.show()

#Univarate visualization of Bearing hecartes across the years using a line plot
coffee_MG.plot(x= 'year', y= 'bearing_hectares')
plt.xlabel('Year')
plt.ylabel('Total Trees')
plt.title('Average Bearing Hectares Over Years')
plt.grid(True)
plt.show()

#Univarate visualization Coffee bags across the years using a line plot
coffee_MG.plot(x= 'year', y= '60kgs_bag')
plt.xlabel('Year')
plt.ylabel('Bag Count')
plt.title('Average 60kg Coffee Bags Over Years')
plt.grid(True)
plt.show()
```

```python
#Distribution of Bearing Trees using a histoplot
sns.histplot(coffee_MG["bearing_trees"],kde=True)
plt.title("Distribution of Bearing Trees")
```

```python
#Distribution of Bearing Hectares using a histoplot
sns.histplot(coffee_MG["bearing_hectares"], kde=True)
plt.title("Distribution of Bearing Hectares")
```

```python
#Distribution of 60kgs Bags using a histoplot
sns.histplot(coffee_MG["60kgs_bag"], kde=True)
plt.title("Distribution of Average 60kg Coffee Bags")
```

```python
#Bivariate visualization Pearson Correlation for reationship between 60kg Bags and Bearing Trees
rel_bag_trees = coffee_MG[["60kgs_bag", "bearing_trees"]].corr()
rel_bag_trees
```

```python
#Bivariate visualization of the relationship between 60kg Bags and Bearing Trees using a scatterplot

sns.scatterplot(x= coffee_MG["60kgs_bag"], y=coffee_MG["bearing_trees"])
plt.xlabel("Number of Bags at 60kg")
plt.ylabel('Bearing Trees')
plt.title('60kg Bags vs Bearing Trees for Minas Gerais')
plt.show()

#Plot linear regression fit 
sns.regplot(data=coffee_MG, x= "60kgs_bag", y="bearing_trees")
plt.xlabel("Number of Bags at 60kg")
plt.ylabel('Bearing Trees')
plt.title('60kg Bags vs Bearing Trees for Minas Gerais')
plt.show()
```

```python
#Bivariate Pearson Correlation for reationship between 60kg Bags and Bearing Heactares
rel_bag_heactares = coffee_MG[["60kgs_bag", "bearing_hectares"]].corr()
rel_bag_heactares

```

```python
#Bivariate visualization of the relationship between 60kg Bags and Bearing Heactares using a scatterplot
sns.scatterplot(x= coffee_MG["60kgs_bag"], y=coffee_MG["bearing_hectares"])
plt.xlabel("Number of Bags at 60kg")
plt.ylabel('Bearing Heactares')
plt.title('60kg Bags vs Bearing Heactares for Minas Gerais')
plt.show()

#Plot linear regression fit 
sns.regplot(data=coffee_MG, x= "60kgs_bag", y="bearing_hectares")
plt.xlabel("Number of Bags at 60kg")
plt.ylabel('Bearing Heactares')
plt.title('60kg Bags vs Bearing Heactares for Minas Gerais')
plt.show()
```
