```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
weather1 = pd.read_csv("../data/weather/weather_data1.csv")
weather2 = pd.read_csv("../data/weather/weather_data2.csv")
```

```python
# TODO: concatenate your data files together and save to a new csv file ("../data/weather/weather_data.csv")
weather = pd.concat([weather1, weather2], axis = 0, join = "inner")
weather.to_csv("../data/weather/weather_data.csv")
```

```python
# Identify the data types for each column within your dataframe
weather.info()
```

```python
# TODO: view first 5 rows of your data
weather.head()
```

```python
# TODO: describe your numeric columns
weather.describe()
```

```python
# TODO: display the null values in your dataframe
weather.isnull().sum()
```

```python
# TODO: BEGIN YOUR EDA!

#Univarate analysis to show relationship between average temperature across the years in Minas Gerais usinf a line graph
# Changed the year data type to a list of string to show as a non-decimal integer in the line graph 
weather.plot(x= 'year', y ='temp_avg')
plt.xlabel('Year')
years = weather.year.tolist()
every_other = years[::2]
plt.xticks(every_other, [str(y) for y in every_other])
plt.ylabel('Average Temperature (C)')
plt.title('Average Temperature Over Years')
plt.grid(True)
plt.show()
```

```python
#Univarate analysis to show relationship between minimum humidity across the years in Minas Gerais
weather.plot(x= 'year', y ='hum_min')
plt.xlabel('Year')
years = weather.year.tolist()
every_other = years[::2]
plt.xticks(every_other, [str(y) for y in every_other])
plt.ylabel('Minimum Humidity')
plt.title('Minimum Humidity Over Years')
plt.grid(True)
plt.show()
```

```python
#Univarate analysis to show relationship between average wind speed across the years in Minas Gerais
weather.plot(x= 'year', y ='wind_avg')
plt.xlabel('Year')
years = weather.year.tolist()
every_other = years[::2]
plt.xticks(every_other, [str(y) for y in every_other])
plt.ylabel('Average Wind')
plt.title('Average Wind Over Years')
plt.grid(True)
plt.show()
```

```python
#Bivariate Pearson Correlation for reationship between maximum rain and humidity 
max_data = weather[["rain_max", "hum_max"]].corr()
max_data
```

```python
#Bivariate Pearson Correlation for reationship between average wind speed and temperature
average_data = weather[["wind_avg", "temp_avg"]].corr()
average_data
```

```python
#Bivariate visualization of the relationship between maximum rain and humidity 

sns.scatterplot(x= weather["rain_max"], y=weather["hum_max"])
plt.xlabel("Maximum Rain")
plt.ylabel('Maximum Humidity')
plt.title('Maximum Rain vs Humidity for Minas Gerais')
plt.show()

#Plot linear regression fit 
sns.regplot(data=weather, x= "rain_max", y="hum_max")
plt.xlabel("Maximum Rain")
plt.ylabel('Maximum Humidity')
plt.title('Maximum Rain vs Humidity for Minas Gerais')
plt.show()

#Bivariate visualization of the relationship between average wind speed and temperature 

sns.scatterplot(x= weather["wind_avg"], y=weather["temp_avg"])
plt.xlabel("Average Wind Speed")
plt.ylabel('Average Temperature')
plt.title('Average Wind Speed vs Temperature for Minas Gerais')
plt.show()

#Plot linear regression fit 
sns.regplot(data=weather, x= "wind_avg", y="temp_avg")
plt.xlabel("Average Wind Speed")
plt.ylabel('Average Temperature')
plt.title('Average Wind Speed vs Temperature for Minas Gerais')
plt.show()
```

```python
#Calculate the maximum rainfall
max_rainfall = weather["rain_max"].max()
max_rainfall
```

```python
#Calculate the year of the maximum rainfall
max_rainfall_year = weather.loc[weather["rain_max"].idxmax(), "year"]
max_rainfall_year
```
